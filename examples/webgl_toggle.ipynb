# WebGL Video Playback

This notebook shows how to use the WebGL rendering path and how to fall back to the native video element.


```python
import VideoPlayerWidget

# Replace with a real video URL (mp4 works best in Chrome/Firefox)
test_url = 'https://iandevlin.github.io/mdn/video-player/video/tears-of-steel-battle-clip-medium.mp4'

# WebGL rendering (default)
v = VideoPlayerWidget.Video()
vid = v.from_url(test_url)
vid
```

```python
# Native video fallback (disable WebGL)
v_native = VideoPlayerWidget.Video(useWebGL=False)
vid_native = v_native.from_url(test_url)
vid_native
```

```python
import VideoPlayerWidget

url = "https://iandevlin.github.io/mdn/video-player/video/tears-of-steel-battle-clip-medium.mp4"
v = VideoPlayerWidget.Video(useWebGL=False)
vid = v.from_url(url)
vid
```
